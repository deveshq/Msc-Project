## Rydberg Atom Dipole Dipole Interaction

Variation of Rydberg atom dipole dipole interaction energy with respect to the sepration between the two atoms:

$$ V_{dd} = \frac{K}{R^3}\big(\vec{d_1}.\vec{d_2} - 3(\vec{d_1}.\hat{R})(\hat{R}.\vec{d_2} )\big)$$

Where $K = \frac{1}{4\pi \epsilon_0}$, $d_i = -n^2ea_0$ dipole moment of atom $i$  ($a_0$: Bohr radius, $n$: Rydberg level), and $R$ is the sepration between two atoms.

```python
import numpy as np
from numpy import cos, pi
from matplotlib import pyplot as plt
```

```python
k = 9 * (10**9) 
a0 = 1  #0.529 * 10**(-10) #Bohr radius
e = 1  #9.1 * 10**(-19) #Electron charge
n = 100 # Rydberg level

#Dipole moment
d = -(n**2)*e*a0 


print('\nThe value of dipole moment :',d)
```

```python
#Sepration between atoms
R = np.arange(1,12,0.01)

alpha = [0,30,45,60,90] #angle between two dipoles
theta = 0 #angle between the d1 and R

V = [0]*len(alpha)

for i in range(len(alpha)):
    V[i] = 10**(-18)*(k/R**3)*((d**2)*cos(alpha[i])
                           - 3*(d**2)*cos(theta)*cos(theta - alpha[i]))
```

```python
V0 = 10**(-18)*(k/R**3)*((d**2)*cos(0)
                           - 3*(d**2)*cos(90)*cos(90))

plt.plot(R,V0)
```

```python
#%matplotlib notebook
fig = plt.figure(figsize=(10,6))

for i in range(len(alpha)):
    plt.plot(R,V[i])

plt.xticks(size=12)
plt.yticks(size=12)
plt.title('\n\n Dipole-Dipole interaction energy VS Atom sepration\n (for different dipole orentations)\n', size=15)
plt.xlabel(r'R$(\mu m) \longrightarrow$', size=15)
plt.ylabel(r'Energy $\longrightarrow$', size=15)
plt.legend(['0','30','45','60','90'])
plt.show()
```

```python
#%matplotlib notebook
theta = [0,30,45,60,90] #angle between the d1 and R

alpha = 56 #angle between two dipoles

V1 = [0]*len(theta)

for i in range(len(theta)):
    V1[i] = 10**(-18)*(k/R**3)*((d**2)*cos(alpha)
                           - 3*(d**2)*cos(theta[i])*cos(theta[i] - alpha))
        
fig = plt.figure(figsize=(10,6))

for i in range(5):
    plt.plot(R,V1[i])

plt.xticks(size=12)
plt.yticks(size=12)
plt.title('\n\n Dipole-Dipole interaction energy VS Atom sepration\n (for different angles b/w dipole & dir. of sepration)\n', size=15)
plt.xlabel(r'R$(\mu m) \longrightarrow$', size=15)
plt.ylabel(r'Energy $\longrightarrow$', size=15)
plt.legend(['0','30','45','60','90'])
plt.show()
```
